#### Resume gráficamente el data set DelayedFlights.csv.

```python
# Importamos las librerías de manipulación y visualización de datos.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargamos el DataFrame usando "pandas"
air_df = pd.read_csv('C:/Users/user1/Downloads/archive/DelayedFlights.csv')
```

#### Crea al menos una visualización por

##### Una variable categòrica (UniqueCarrier)
Visualizamos la variable categórica frete a DepDelay, por ejemplo.

```python
# Graficamos usando, en este caso, Seaborn.
sns.stripplot(x="UniqueCarrier", y="DepDelay", data=air_df)
```

##### Una variable numèrica (ArrDelay)
Visualizamos el retraso de las llegadas según la distancia del vuelo

```python
# Podemos hacer un gráfico de dispersión usando una variable categórica, con matplotlib
plt.scatter(air_df.ArrDelay,air_df.Distance,s=1)
plt.xlabel("Arrival Delay")
plt.ylabel("Distance")
plt.show()
```

##### Una variable numèrica i una categòrica (ArrDelay i UniqueCarrier)
Creamos un subarray de las variables UniqueCarrier y ArrDelay para calcular la media de retraso en la llegada de cada compañía.

```python
# Creamos el subarray y calculamos la media de la compañía.
carrier_delay = air_df[["UniqueCarrier", "ArrDelay"]]
mean_cd = carrier_delay.groupby("UniqueCarrier").mean()

# Mostramos el gráfico
plt.bar(mean_cd.index,mean_cd.ArrDelay)
plt.show()
```

##### Dos variables numéricas (ArrDelay y DepDelay)

```python
sns.stripplot(x="DepDelay", y="ArrDelay", data=air_df)
```

##### Tres variables (ArrDelay, DepDelay i UniqueCarrier)

```python
sns.lmplot(x='ArrDelay', y='DepDelay', data=air_df, hue='UniqueCarrier', fit_reg=False)
```

##### Más de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier).

```python
sns.scatterplot(data = air_df, x = "ArrDelay", y = "DepDelay", hue = "UniqueCarrier", size = "AirTime")
```
